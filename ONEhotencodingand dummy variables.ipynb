```python
#convert categorical into numerical format
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
tips=sns.load_dataset("tips")
print(tips)
```

```python
#here we have 4 categorical varaibles sex,smoker,day,time ,, target variables:total-bill

dummy_df=pd.get_dummies(tips)
dummy_df    #now shape has been changed and there is dummy variables created 
#sex have 2 classes ,so 2 dummy variables created.#we have 2 columns but we can arrange it in 1 column 
```

```python
dummy_df.columns
```

```python
#1 dummy variables is enough to maintain data ,drop_first=False
pd.get_dummies(tips,drop_first=True)   #drop_first=True will arrange all dummy columns in a better way 
#using this we can ML mmodel
```

# One-hot encoding with scikit learn 


```python
from sklearn.preprocessing import OneHotEncoder  #onehotencoder  class called here
onh=OneHotEncoder(sparse=False)

#drop=we have to remove 1 class column  bydefault 
#datatype=float64
```

```python
array_format=onh.fit_transform(tips[['sex','smoker','day','time']])
array_format  #we got numpy array format data and convert into dataframe 
```

```python
on_hoc_df=pd.DataFrame(array_format,columns=['sex_Male', 'sex_Female', 'smoker_Yes',
       'smoker_No', 'day_Thur', 'day_Fri', 'day_Sat', 'day_Sun', 'time_Lunch',
       'time_Dinner'])
```
